```python
import os
import json
import random
import supervisely as sly
from tqdm import tqdm
from fastai.vision.all import *
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from pycocotools.coco import COCO

input_path = "./windturbine"
output_path = "./results"
os.makedirs(output_path, exist_ok=True)

# Creating Supervisely project from local directory.
project = sly.Project(input_path, sly.OpenMode.READ)
print("Opened project: ", project.name)
print("Number of images in project:", project.total_items)

# Showing annotations tags and classes.
print(project.meta)

# Iterating over classes in project, showing their names, geometry types and colors.
for obj_class in project.meta.obj_classes:
    print(
        f"Class '{obj_class.name}': geometry='{obj_class.geometry_type}', color='{obj_class.color}'",
    )

# Iterating over tags in project, showing their names and colors.
for tag in project.meta.tag_metas:
    print(f"Tag '{tag.name}': color='{tag.color}'")

print("Number of datasets (aka folders) in project:", len(project.datasets))

progress = tqdm(project.datasets, desc="Processing datasets")

# Function to draw bounding boxes on the image
def draw_bounding_boxes(img, annotations, output_img_path):
    fig, ax = plt.subplots(1)
    ax.imshow(img)
    
    for label in annotations.labels:
        bbox = label.geometry.to_bbox()
        rect = patches.Rectangle((bbox.left, bbox.top), bbox.width, bbox.height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
    
    plt.savefig(output_img_path)
    plt.close()

# Iterating over images in dataset, using the paths to the images and annotations.
for dataset in project.datasets:
    for item_name, image_path, ann_path in dataset.items():
        print(f"Item '{item_name}': image='{image_path}', ann='{ann_path}'")

        ann_json = json.load(open(ann_path))
        ann = sly.Annotation.from_json(ann_json, project.meta)

        img = sly.image.read(image_path)  # rgb - order

        # Draw bounding boxes and save the image
        res_image_path = os.path.join(output_path, item_name)
        draw_bounding_boxes(img, ann, res_image_path)

        progress.update(1)

# Load COCO annotations
coco = COCO('annotation.json')

# Retrieve category information
category_ids = coco.loadCats(coco.getCatIds())
print(category_ids)

# Generate a list of all image IDs
image_ids = coco.getImgIds(catIds=coco.getCatIds())

# Select a random image ID
random_image_id = random.choice(image_ids)

# Load image metadata
img_metadata = coco.loadImgs(random_image_id)[0]

# Load and display the image with annotations
def display_image_with_annotations(img_metadata, coco):
    image_path = os.path.join(input_path, 'train', img_metadata['file_name'])
    img = plt.imread(image_path)
    
    # Load annotations
    ann_ids = coco.getAnnIds(imgIds=img_metadata['id'], catIds=coco.getCatIds(), iscrowd=None)
    anns = coco.loadAnns(ann_ids)
    
    fig, ax = plt.subplots(1)
    ax.imshow(img)
    
    # Draw bounding boxes
    for ann in anns:
        bbox = ann['bbox']
        rect = patches.Rectangle((bbox[0], bbox[1], bbox[2], bbox[3]), linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
    
    plt.axis('off')  # Hide axes for better visualization
    plt.show()

# Display a random image with annotations
display_image_with_annotations(img_metadata, coco)
```
